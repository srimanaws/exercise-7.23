```python
#Exercise 7.23 by sriman

import pandas as pd

# a) Create a DataFrame named temperatures
data = {
    'Maxine': [72, 75, 68],
    'James': [70, 78, 65],
    'Amanda': [68, 74, 71]
}

temperatures = pd.DataFrame(data)
print("Part a: DataFrame temperatures:\n", temperatures)

# b) Recreate the DataFrame with custom indices
temperatures = pd.DataFrame(data, index=['Morning', 'Afternoon', 'Evening'])
print("\nPart b: DataFrame with custom indices:\n", temperatures)

# c) Select the column for 'Maxine'
maxine_temps = temperatures['Maxine']
print("\nPart c: Maxine's temperature readings:\n", maxine_temps)

# d) Select the row of 'Morning' temperature readings
morning_temps = temperatures.loc['Morning']
print("\nPart d: Morning temperature readings:\n", morning_temps)

# e) Select rows for 'Morning' and 'Evening'
morning_evening_temps = temperatures.loc[['Morning', 'Evening']]
print("\nPart e: Morning and Evening temperature readings:\n", morning_evening_temps)

# f) Select columns for 'Amanda' and 'Maxine'
amanda_maxine_temps = temperatures[['Amanda', 'Maxine']]
print("\nPart f: Amanda and Maxine's temperature readings:\n", amanda_maxine_temps)

# g) Select elements for 'Amanda' and 'Maxine' in 'Morning' and 'Afternoon'
specific_temps = temperatures.loc[['Morning', 'Afternoon'], ['Amanda', 'Maxine']]
print("\nPart g: Amanda and Maxine's temperatures in Morning and Afternoon:\n", specific_temps)

# h) Use the describe method for descriptive statistics
stats = temperatures.describe()
print("\nPart h: Descriptive statistics of temperatures:\n", stats)

# i) Transpose temperatures
transposed_temperatures = temperatures.T
print("\nPart i: Transposed temperatures:\n", transposed_temperatures)

# j) Sort temperatures by column names in alphabetical order
sorted_temperatures = temperatures.reindex(sorted(temperatures.columns), axis=1)
print("\nPart j: Sorted temperatures by column names:\n", sorted_temperatures)
```
